```python
from youtube_transcript_api import YouTubeTranscriptApi

# retrieve the available transcripts
transcript_list = YouTubeTranscriptApi.list_transcripts('hPt4N1iVtlU')

# open a file to store the transcript information
with open('transcript_info.txt', 'w', encoding='utf-8') as file:
    # iterate over all available transcripts
    for transcript in transcript_list:
        # write transcript metadata to the file
        file.write(f"Video ID: {transcript.video_id}\n")
        file.write(f"Language: {transcript.language}\n")
        file.write(f"Language Code: {transcript.language_code}\n")
        file.write(f"Is Generated: {transcript.is_generated}\n")
        file.write(f"Is Translatable: {transcript.is_translatable}\n")
        file.write(f"Translation Languages: {transcript.translation_languages}\n")
        
        # fetch the actual transcript data and write to the file
        file.write("Transcript:\n")
        file.write(str(transcript.fetch()) + "\n")
        
        # translating the transcript will return another transcript object
        translated_transcript = transcript.translate('en').fetch()
        file.write("Translated Transcript:\n")
        file.write(str(translated_transcript) + "\n")
        
        file.write("\n")  # add a newline between each transcript entry

print("Transcript information saved to transcript_info.txt")

# open the file and read its contents
with open('transcript_info.txt', 'r', encoding='utf-8') as file:
    transcript_info = file.read()
    print("Contents of transcript_info.txt:\n")
    print(transcript_info)
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
from chat_downloader import ChatDownloader

# Function to format the chat message
def format_chat_message(message):
    return f"{message['timestamp']} [{message['author']['name']}]: {message['message']}"

# Retrieve the available transcripts
transcript_list = YouTubeTranscriptApi.list_transcripts('hPt4N1iVtlU')

# Open a file to store the transcript information and chat messages
with open('transcript_and_chat_info.txt', 'w', encoding='utf-8') as file:
    # Iterate over all available transcripts
    for transcript in transcript_list:
        # Write transcript metadata to the file
        file.write(f"Video ID: {transcript.video_id}\n")
        file.write(f"Language: {transcript.language}\n")
        file.write(f"Language Code: {transcript.language_code}\n")
        file.write(f"Is Generated: {transcript.is_generated}\n")
        file.write(f"Is Translatable: {transcript.is_translatable}\n")
        file.write(f"Translation Languages: {transcript.translation_languages}\n")
        
        # Fetch the actual transcript data and write to the file
        file.write("Transcript:\n")
        file.write(str(transcript.fetch()) + "\n")
        
        # Translating the transcript will return another transcript object
        translated_transcript = transcript.translate('en').fetch()
        file.write("Translated Transcript:\n")
        file.write(str(translated_transcript) + "\n")
        
        file.write("\n")  # Add a newline between transcript entries

    # Get chat messages from the video
    url = 'https://www.youtube.com/watch?v=hPt4N1iVtlU'
    chat = ChatDownloader().get_chat(url)
    
    # Write chat messages to the file
    file.write("Chat Messages:\n")
    for message in chat:
        formatted_message = format_chat_message(message)
        file.write(formatted_message + "\n")

print("Transcript and chat information saved to transcript_and_chat_info.txt")
   
```

```python
with open('transcript_info.txt', 'r', encoding='utf-8') as file:
    transcript_info = file.read()
    print("Transcript and chat information saved to transcript_and_chat_info.txt:\n")
    print(transcript_info)
```
